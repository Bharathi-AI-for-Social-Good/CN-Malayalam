```python
from textblob import TextBlob

# Function to get sentiment polarity (for text with some English or transliterated Malayalam)
def get_sentiment(text):
    try:
        return TextBlob(str(text)).sentiment.polarity
    except:
        return 0.0

# Apply to dataset
df['H/T_Sentiment'] = df['H/T'].apply(get_sentiment)
df['CS_Sentiment'] = df['CS'].apply(get_sentiment)

# Plot sentiment distribution
plt.figure(figsize=(12, 6))
sns.histplot(df['H/T_Sentiment'], bins=30, color='red', label='H/T Sentiment', kde=True, stat="density", alpha=0.6)
sns.histplot(df['CS_Sentiment'], bins=30, color='green', label='CS Sentiment', kde=True, stat="density", alpha=0.6)
plt.title("Sentiment Polarity Distribution")
plt.xlabel("Sentiment Polarity (-1 = Negative, +1 = Positive)")
plt.ylabel("Density")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Calculate lengths of H/T and CS content
df['H/T_Length'] = df['H/T'].astype(str).apply(len)
df['CS_Length'] = df['CS'].astype(str).apply(len)

# Plot histograms
plt.figure(figsize=(14, 6))

# Histogram for H/T content length
plt.subplot(1, 2, 1)
plt.hist(df['H/T_Length'], bins=30, color='darkred', alpha=0.8)
plt.title("Length Distribution of H/T Content")
plt.xlabel("Character Count")
plt.ylabel("Frequency")

# Histogram for CS content length
plt.subplot(1, 2, 2)
plt.hist(df['CS_Length'], bins=30, color='darkgreen', alpha=0.8)
plt.title("Length Distribution of Counter-Speech (CS)")
plt.xlabel("Character Count")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
from PIL import Image, ImageDraw, ImageFont

# Reprocess CS content by category with stopword removal explicitly enforced
cs_images_cleaned = []
top_4_categories = df['Category'].value_counts().nlargest(4).index
malayalam_font_path = "NotoSansMalayalam-Bold.ttf"

def create_malayalam_bar_chart_image_with_spacing(data, font_path, image_size=(1000, 460), bar_color="teal"):
    img = Image.new("RGB", image_size, color="white")
    draw = ImageDraw.Draw(img)
    font = ImageFont.truetype(font_path, size=26)
    title_font = ImageFont.truetype(font_path, size=34)

    # Draw title
    draw.text((10, 10), data['Category'].iloc[0], font=title_font, fill="black")

    # Limit to 6 rows
    data = data.head(6)

    max_freq = data['Frequency'].max()
    bar_max_width = image_size[0] - 350
    top = 70
    bar_height = 30
    spacing = 30  # more vertical space, including 7th row space

    for i, row in data.iterrows():
        y = top + i * (bar_height + spacing)
        bar_width = int((row['Frequency'] / max_freq) * bar_max_width)
        draw.text((10, y), row['Word'], font=font, fill="black")
        draw.rectangle([280, y, 280 + bar_width, y + bar_height], fill=bar_color)
        draw.text((290 + bar_width, y), str(row['Frequency']), font=font, fill="black")

    return img

for category in top_4_categories:
    cs_text = " ".join(df[df['Category'] == category]['CS'].dropna().astype(str).tolist())

    # Tokenize and clean with stopword removal
    tokens = re.findall(r'[\u0D00-\u0D7F]+', cs_text)
    filtered_tokens = [word for word in tokens if word not in malayalam_stopwords and len(word) > 1]

    # Get top words
    word_freq = Counter(filtered_tokens).most_common(10)
    cat_data = pd.DataFrame(word_freq, columns=['Word', 'Frequency'])
    cat_data['Category'] = category

    # Generate image
    img = create_malayalam_bar_chart_image_with_spacing(cat_data, malayalam_font_path)
    cs_images_cleaned.append(img)

# Combine into 2x2 grid
cs_row1_cleaned = Image.new("RGB", (2000, 460))
cs_row2_cleaned = Image.new("RGB", (2000, 460))
cs_row1_cleaned.paste(cs_images_cleaned[0], (0, 0))
cs_row1_cleaned.paste(cs_images_cleaned[1], (1000, 0))
cs_row2_cleaned.paste(cs_images_cleaned[2], (0, 0))
cs_row2_cleaned.paste(cs_images_cleaned[3], (1000, 0))

cs_grid_cleaned = Image.new("RGB", (2000, 920))
cs_grid_cleaned.paste(cs_row1_cleaned, (0, 0))
cs_grid_cleaned.paste(cs_row2_cleaned, (0, 460))




# Plot both H/T and CS length distributions by category side by side

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 6), sharey=True)

# H/T Length by Category
sns.boxplot(ax=axes[0], data=df, x='Category', y='H/T_Length', palette="Set2")
axes[0].set_title("H/T Content Length by Category")
axes[0].set_xlabel("Category")
axes[0].set_ylabel("Length (characters)")
axes[0].tick_params(axis='x', rotation=45)

# CS Length by Category
sns.boxplot(ax=axes[1], data=df, x='Category', y='CS_Length', palette="Set3")
axes[1].set_title("Counter-Speech Length by Category")
axes[1].set_xlabel("Category")
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import seaborn as sns
import re
from collections import Counter

# Basic Malayalam stopword list (can be expanded)
malayalam_stopwords = {
    'ഒരു', 'എന്ന്', 'ഇത്', 'അത്', 'ആണ്', 'ഉണ്ട്', 'ഇല്ല', 'വീട്', 'വരെ', 'കൂടി', 'എന്നും', 'പോലെ', 'ഇനി',
    'അവൻ', 'അവൾ', 'അവർ', 'നമുക്ക്', 'ഞാൻ', 'നീ', 'നിനക്ക്', 'ഇവ', 'ഉണ്ടായിരിക്കും', 'ഉണ്ടാകില്ല',
    'ചെയ്യുക', 'ചെയ്തു', 'ഇരിക്കുന്നു', 'ആകുന്നു', 'ഉള്ളത്'
}

malayalam_font_path = "NotoSansMalayalam-Regular.ttf"

text_ht = " ".join(df['H/T'].dropna().astype(str).tolist())

# Basic tokenizer for Malayalam: split on non-Malayalam characters
tokens = re.findall(r'[\u0D00-\u0D7F]+', text_ht)

# Remove stopwords
filtered_tokens = [word for word in tokens if word not in malayalam_stopwords and len(word) > 1]

# Count frequency of remaining words
word_freq = Counter(filtered_tokens)

# Generate word cloud from filtered, tokenized, cleaned text
wordcloud_cleaned = WordCloud(font_path=malayalam_font_path, width=800, height=400,
                              background_color='white').generate_from_frequencies(word_freq)

#plot
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud_cleaned, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud of Original H/T Content", fontsize=16)
plt.show()
```
